```python
ht = ht.filter(ht.transcript=="ENST00000366684").persist()
```

```python
ht_2 = ht.annotate(last2=hl.scan.take(ht.cumulative_exp,2,ordering=-ht.cumulative_exp))
```

```python
ht_2 = ht.annotate(last2=hl.scan.take(ht.cumulative_exp,1,ordering=-ht.cumulative_exp))
```

```python
ht_2.show()
```

```python
ht_2 = ht_2.annotate(last1 = ht_2.last2.first())
```

```python
ht_2.show()
```

```python
ht_2 = ht_2.annotate(exp = ht_2.cumulative_exp - hl.coalesce(ht_2.last1,0))
```

```python
ht_2.show()
```

```python
ht_2.aggregate(hl.agg.sum(ht_2.exp))
```

```python
ht_2.aggregate(hl.agg.sum(ht_2.exp)-hl.agg.take(ht_2.total_exp, 1))
```
